```python
COMPETITION_ID = 9
SEASON_ID = 281
```

```python
all_matches = sb.matches(competition_id=COMPETITION_ID, season_id=SEASON_ID)
all_matches
```

```python
import pandas as pd

wirtz_match_logs = pd.read_csv('florian_wirtz_all_match_logs_23_24.csv')
matches_wirtz_scores = wirtz_match_logs[wirtz_match_logs["Performance_Gls"] > 0]
matches_wirtz_scores
```

```python
match_dates = matches_wirtz_scores["Date"]
match_dates
```

```python
# select matches with Bayer Leverkusen participating and the date is in match_dates
all_matches_wirtz_scores = all_matches[(all_matches['home_team'] == 'Bayer Leverkusen') | (all_matches['away_team'] == 'Bayer Leverkusen')][all_matches['match_date'].isin(match_dates)]
all_matches_wirtz_scores
```

```python
MATCH_IDS = all_matches_wirtz_scores['match_id']
MATCH_IDS
```

```python
sb.lineups(match_id=MATCH_IDS[0])["Bayer Leverkusen"]
```

```python
PLAYER_ID = 40724
```

```python
from pandas import DataFrame

# get all goal events for the given match and player
shot_events = DataFrame(sb.events(match_id=MATCH_IDS[0], split=True, flatten_attrs=True, fmt="dataframe")["shots"])
shot_events_wirtz = shot_events[shot_events["player_id"] == PLAYER_ID]
goal_events_writz = shot_events_wirtz[shot_events_wirtz["shot_outcome"] == "Goal"]
goal_events_writz
```

```python
GOAL_EVENTS_WIRTZ_IDS = goal_events_writz["id"].tolist()
GOAL_EVENTS_WIRTZ_IDS
```

```python
import pandas as pd
# load statsbomb 360 data
match_360_df = pd.read_json(f"https://raw.githubusercontent.com/statsbomb/open-data/master/data/three-sixty/{MATCH_IDS[0]}.json")
match_360_goals = match_360_df[match_360_df["event_uuid"].isin(GOAL_EVENTS_WIRTZ_IDS)]
match_360_goals
```

```python
merged_df = pd.merge(goal_events_writz, match_360_goals, left_on="id", right_on="event_uuid")
merged_df
```

```python
# lets put it all together
def get_goal_events_wirtz(match_id, player_id):
    shot_events = DataFrame(sb.events(match_id=match_id, split=True, flatten_attrs=True, fmt="dataframe")["shots"])
    shot_events_wirtz = shot_events[shot_events["player_id"] == player_id]
    goal_events_writz = shot_events_wirtz[shot_events_wirtz["shot_outcome"] == "Goal"]
    return goal_events_writz

def get_match_360_goals(match_id, goal_events_wirtz):
    match_360_df = pd.read_json(f"https://raw.githubusercontent.com/statsbomb/open-data/master/data/three-sixty/{match_id}.json")
    match_360_goals = match_360_df[match_360_df["event_uuid"].isin(goal_events_wirtz["id"].tolist())]
    return match_360_goals
```

```python
# iterate over all matches and in each match merge goal events with 360 data
all_goals = DataFrame()
for match_id in MATCH_IDS:
    goal_events_writz = get_goal_events_wirtz(match_id, PLAYER_ID)
    match_360_goals = get_match_360_goals(match_id, goal_events_writz)
    merged_df = pd.merge(goal_events_writz, match_360_goals, left_on="id", right_on="event_uuid")
    # concatenate all dataframes
    all_goals = pd.concat([all_goals, merged_df], axis=0, ignore_index=True)
    print(all_goals.shape)

all_goals
```

```python
all_goals.to_json("florian_wirtz_all_goal_events.json", index=False)
```
