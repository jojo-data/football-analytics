```python
import pandas as pd
```

```python
df = pd.read_html("https://fbref.com/en/players/e7fcf289/matchlogs/2023-2024/Florian-Wirtz-Match-Logs", attrs={"id": "matchlogs_all"})
```

```python
df[0].columns
```

```python
df[0].head()
```

```python
# clean up the column names
df[0].columns = [col[0] + '_' + col[1] if not col[0].startswith('Unnamed') else col[1] for col in df[0].columns]
df[0].columns
```

```python
df[0].head()
```

```python
# Drop rows where Date is NaN
df[0].dropna(subset=['Date'], inplace=True)
```

```python
# remove rows where Min is not a number
df[0] = df[0][df[0]['Min'].str.isnumeric()]
```

```python
df[0].to_csv("florian_wirtz_all_match_logs_23_24.csv", index=False)
```
